```python
import pandas as pd
import csv
import matplotlib.pyplot as plt

df = pd.read_csv("star_with_gravity.csv")
df.head()
```

```python
radius = df['Radius'].to_list()
mass = df['Mass'].to_list()
gravity = df['Gravity'].to_list()

planet_details = [radius,mass,gravity]
planet_details.sort()
```

```python
plt.plot(radius,color ='purple',  label ='Radius')
plt.plot(mass,color ='pink', label ='Mass')
plt.xlabel('Mass')
plt.ylabel('Radius')

plt.title('The Relationship Between the Mass and Radius of Stars')

plt.legend()
plt.show()
```

```python
plt.plot(gravity,color ='blue',  label ='Gravity')
plt.plot(mass,color ='pink', label ='Mass')
plt.xlabel('Mass')
plt.ylabel('Radius')

plt.title('The Relationship Between the Mass and Gravity of Stars')

plt.legend()
plt.show()
```
